```python
import sys
sys.path.append('../')
from src.society import Society
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
n = 1000
p = (0.32, 0.24, 0.27, 0.13)
_max = 2000
g = []
a = []
b = []
o = []
ab = []
af = []
bf = []
of = []
s = Society(n, p, _max)

print(s)
```

```python
for _ in range(1000):
    s.next_gen()
    f_p, m_p, a_p, b_p, o_p, ab_p = s.get_proportion()
    af_p, bf_p, of_p = s.get_blood_feature_portion()
    g.append(s.gen)
    a.append(a_p)
    b.append(b_p)
    o.append(o_p)
    ab.append(ab_p)
    af.append(af_p)
    bf.append(bf_p)
    of.append(of_p)
    if s.gen % 100 == 0:
        print(s)

plt.figure()
plt.title('Blood Type Proportion')
plt.xlabel('Generation')
plt.ylabel('Portion')
plt.plot(g, a, label='A')
plt.plot(g, b, label='B')
plt.plot(g, o, label='O')
plt.plot(g, ab, label='AB')
plt.legend()
plt.show()
plt.figure()
plt.title('ABO DNA pool Proportion')
plt.xlabel('Generation')
plt.ylabel('Portion')
plt.plot(g, af, label='A')
plt.plot(g, bf, label='B')
plt.plot(g, of, label='O')
plt.legend()
plt.show()
```
